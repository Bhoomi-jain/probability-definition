```python
import pandas as pd
import numpy as np
import scipy.stats as s
import matplotlib.pyplot as plt
```

```python
coin_tosses = np.random.binomial(n=1,p=0.5,size=10000000)
```

```python
coin_tosses
```

```python
frequency = np.count_nonzero(coin_tosses)
```

```python
relative_frequency = frequency/coin_tosses.shape[0]
```

```python
relative_frequency
```

```python
number_of_tosses = 15
```

```python
k = np.arange(start=0,stop=number_of_tosses+1)
```

```python
k
```

```python
p_y_equals_k = s.binom.pmf(n=number_of_tosses,p=0.35,k=k)
```

```python
p_y_equals_k
```

```python
print(np.sum(p_y_equals_k))
```

```python
plt.bar(x=k,height=p_y_equals_k)
plt.xlabel("Number of Desired outcomes out of 5 Coin tosses")
plt.ylabel("Probability of getting k desired outcomes out of 5 Coin tosses")
```
